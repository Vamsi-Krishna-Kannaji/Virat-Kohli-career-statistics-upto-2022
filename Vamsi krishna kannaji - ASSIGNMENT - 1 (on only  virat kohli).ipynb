```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(7, 6))

ax1.fill_between(df['Years'],df['lOOs'])
ax1.set_title("fill between Years and 100's")

ax2.fill_between (df['Years'],df['lOOs'], 1)
ax2.set_title("fill between Years and 100's")

ax3.fill_between(df['Years'],df['lOOs'], df['Runs'])
ax3.set_title("fill between Years and Runs")
ax3.set_xlabel('x')
fig.tight_layout()
```

```python
fig, ax=plt.subplots()
bar_lables =['red','blue','green','orange']
bar_colours=['tab:red','tab:blue','tab:green','tab:orange']
ax.bar(df['Years'],df['Inns'], label= bar_lables ,color=bar_colours)
ax.set_ylabel('Inns')
ax.set_title("Inns played in a year")
ax.legend_title=("Runs scored")
plt.show()
```

```python
fig, ax=plt.subplots()
bar_lables=['red','blue','green','orange']
bar_colours=['tab:red','tab:blue','tab:green','tab:orange']
ax.bar(df['Years'],df['Runs'], label= bar_lables ,color=bar_colours)
ax.set_ylabel('Runs')
ax.set_title("Runs scored in year")
ax.legend_title=("Runs scored")
plt.show()
```

```python
fig, ax=plt.subplots()
bar_lables=['red','blue','green','orange']
bar_colours=['tab:red','tab:blue','tab:green','tab:orange']
ax.bar(df['Years'],df['lOOs'], label=bar_lables,color=bar_colours)
ax.set_ylabel('lOOs')
ax.set_title("lOOs scored in year")
ax.legend_title=("lOOs scored")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df['Years']
fig1, ax1 = plt.subplots()
ax1.pie(df['Runs'], labels=labels, autopct='%1.0f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df['Years']
fig1, ax1 = plt.subplots()
ax1.pie(df['50s'], labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
